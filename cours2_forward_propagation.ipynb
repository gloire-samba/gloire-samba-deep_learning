```python
import numpy as np
import pandas as pd
```

```python
# Simulé les données

np.random.seed(42)
num_sample = 10
```

```python
age = np.random.randint(18, 90, num_sample)
age
```

```python
income = np.random.randint(20000, 100000, num_sample)
income
```

```python
loan_amount = income = np.random.randint(50000, 100000, num_sample)
```

```python
num_acounts = np.random.randint(1, 10, num_sample)
```

```python
# création des caractéristique d'entrée
x = np.column_stack((age, income, loan_amount, num_acounts))
x
```

```python
x = np.column_stack((age, income, loan_amount, num_acounts)).T
x
```

```python
# Fonction d'activation Relu
def relu(x):
  return np.maximum(0,x)

# fonction d'activation sigmoide
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
# Nombre de neurones
hidden_size = 5
np.random.seed(42)
w1 = np.random.rand(hidden_size, x.shape[0])
w1
```

```python
b1 = np.random.rand(hidden_size,1)
b1
```

```python
output_size  = 1
w2 = np.random.rand(output_size, hidden_size)
b2 = np.random.rand(output_size,1)
```

```python
# forward pass
z1 = np.dot(w1,x) + b1
z1
```

```python
a1 = relu(z1)
a1
```

```python
z2 = np.dot(w2,a1) + b2
z2
```

```python
a2 = sigmoid(z2)
a2
```
